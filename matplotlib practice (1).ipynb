# Scatter Plot

```python
flowers_df = sns.load_dataset("iris")
flowers_df
```

```python
flowers_df.species.unique()
```

```python
flowers_df["species"].unique()
```

```python
plt.plot(flowers_df.sepal_length, flowers_df.sepal_width);
```

```python
#The output is not very informative as there are too many combinations of the two properties within the dataset. 
#There doesn't seem to be simple relationship between them.

#We can use a scatter plot to visualize how sepal length & sepal width vary using 
#the scatterplot function from the seaborn module (imported as sns).
```

```python
sns.scatterplot(x=flowers_df.sepal_length,y=flowers_df.sepal_width);
```

```python
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width, hue=flowers_df.species, s=10);
```

```python
plt.figure(figsize=(12,6))
plt.title('Sepal Dimensions')
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width, hue=flowers_df.species, s=100);

# if you give the figure command after the scatter plot the figsize won't change
```

```python
plt.hist(flowers_df.sepal_width);
```

```python
# the above histogram tells that what are the number of flowers within a range of sepal width. 
# For ex. the maximum numbers of flower have sepal width between 3 to 3.25                        
```

```python
plt.hist(flowers_df.sepal_width,bins=5);
```

```python
import numpy as np
# Specifying the boundaries of each bin
plt.hist(flowers_df.sepal_width, bins=np.arange(2,5,0.25));
```

```python
# Bins of unequal sizes
plt.hist(flowers_df.sepal_width, bins=[1, 3, 4, 4.5]);
```

```python
#separate histograms for each species

setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
plt.hist(setosa_df.sepal_width, alpha=0.5, bins=np.arange(2, 5, 0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.7, bins=np.arange(2, 5, 0.25));

# here alpha represents opacity of both species
```

```python
# using stacked histogrmas

import numpy as np
plt.figure(figsize=(12,6))
plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], bins=np.arange(2,5,0.25)
        ,stacked=True);

plt.title('count of flowers as per sepal width');
plt.xlabel('sepal width');
plt.ylabel('count of flowers');
plt.legend(['setosa','versicolor','virginica']);
```

```python
tips_df = sns.load_dataset("tips");
tips_df
```

```python
import pandas as pd
average_bill=tips_df.groupby('day')[['total_bill']].mean()

average_bill.rename(columns={'total_bill':'mean_bill'},inplace = True)
average_bill
```

```python
plt.bar(average_bill.index,average_bill.mean_bill);
```

```python
# the above graph shows the average bill for each day
# but it involves two steps , first we have to use group by and make a new dataframe then we have to make the bar chart
# seaborn library provides a readymade feature  to do so, so this task can be completed in a single step
```

```python
sns.barplot(x='day',y='total_bill',data=tips_df );
```

```python
# The lines cutting each bar represent the amount of variation in the values. 
# For instance, it seems like the variation in the total bill is relatively high on Fridays and low on Saturday.
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df);
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='total_bill', y='day', hue='sex', data=tips_df);
```

# HeatMap

```python
#A heatmap is used to visualize 2-dimensional data like a matrix or a table using colors.
```

```python
flights_df=sns.load_dataset('flights').pivot('month','year','passengers')
flights_df
```

```python
# the figure function should always be given on the top otherwise it will generate an error.
plt.figure(figsize=(12,6))
plt.title("No. of Passengers (1000s)");
sns.heatmap(flights_df);
```

```python
plt.figure(figsize=(12,6))
plt.title("No. of Passengers (1000s)");
sns.heatmap(flights_df,fmt='d',annot=True,cmap='Blues');
# here the annot=True is used to display the actual values on the map
# cmap is used to give the color pattern of the heatmap
# if you don't give the fmt=d command, the numbers will come in scientific notation.
```
